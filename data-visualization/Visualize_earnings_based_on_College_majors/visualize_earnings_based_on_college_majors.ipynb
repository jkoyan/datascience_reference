Using Visualizations, we can start to explore questions from the dataset like:

#### Explore the data

- Generate scatter plots in separate jupyter notebook cells to explore the following relations:
  - Sample_size and Median
  - Sample_size and Unemployment_rate
  - Full_time and Median
  - ShareWomen and Unemployment_rate
  - Men and Median

- Use the plots to explore the following questions:
  - Do students in more popular majors make more money?
  - Do students that majored in subjects that were majority female make more money?

````python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='Scatter', title='Median vs. Sample size')
````

````python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="Scatter", title="Unemployment rate vs. Sample size")
````

````python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="Scatter", title="Unemployment rate vs. Share woman")
````

````python
recent_grads.plot(x="Men", y="Median", kind="Scatter", title="Median vs. Men")
````

````python
recent_grads.plot(x="Women", y="Median", kind="Scatter", title="Median vs. Women")
````

#### Do students in more popular majors make more money? 

- From the Median vs. Men and the Median vs. Women plot it is clear that students from the less popular major categories make more money, the data is concentrated towards the left side, which is the less popular ones since there are fewer people taking those majors, In cpntrast the salary range is between 20K to 150K (from the Median vs. Women plot). As we scan the plot to the right the salary range narrows down between 30K to 50K, and it narrows even more to the right.

#### Do students that majored in subjects that were majority female make more money?
To answer this question let's plot the ShareWoman vs. the Median scatter plot. If the Median salary of a category where the Share of Women is higher than we can conclude by saying yes. Let's see -

````python
recent_grads.plot(x="ShareWomen", y="Median", kind="Scatter", title="Median vs. ShareWomen")
````

From the Median vs. ShareWoman plot we can conlude that the data paints a different picture, the median salary is in the lower range for Majors that have majority women.

#### Is there any link between the number of full-time employees and median salary?
Let's plot the Median vs. Fulltime plot

````python
ax = recent_grads.plot(x='Full_time', y='Median', kind='Scatter', title='Median vs. Full time')
````

There does appear to be a negative corelation between the number of full time employees with a Major and the Median Salary, as the number of full time employees increases the Median salary slides down the slope. However this needs to be investigated within the context of the Major, that is going back to the original question if more popular Majors make more money and the answer is negative. 

#### Explore normal distributions using Histograms 
In the next cells we will plot histograms to explore distributions for the following columns -
- Sample_size
- Median
- Employed
- Full_time
- ShareWomen
- Unemployment_rate
- Men
- Women

we will be using the Series.hist method to generate the histogram -
```
recent_grads['Median'].hist(bins=20, range=(10000, 200000))
```

````python
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
fig = plt.figure(figsize=(10,20))
for r in range(1,9):
    ax = fig.add_subplot(4,2, r)
    ax = recent_grads[cols[r-1]].plot(kind="hist", rot=40, title="hist - {}".format(cols[r-1]))    
plt.show()
#recent_grads['Median'].hist(bins=25, range=(10000, 150000))
````

#### Scatter Matrix plot

A scatter matrix plot combines both the scatter and histogram plots into one grid of plots and allows us to explore potential relationships and distributions simultaneously. Because scatter matrix plots are frequently used in the exploratory data analysis stage, pandas has a scatter_matrix method made available that generates the plot for us. The method is part of the pandas.tools.plotting module and needs to be imported separately.

Instructions for the next few cells -
- Import scatter_matrix from pandas.tools.plotting
- Create a 2 by 2 scatter matrix plot using the Sample_size and Median columns.
- Create a 3 by 3 scatter matrix plot using the Sample_size, Median, and Unemployment_rate columns.
- Explore the questions from the last few steps using these scatter matrix plots. You may need to create more scatter matrix plots.

````python
from pandas.plotting import scatter_matrix 
cols = ['Sample_size', 'Median', 'Unemployment_rate']
scatter_matrix(recent_grads[cols[0:2]], figsize=(6,6))
````

````python
scatter_matrix(recent_grads[cols[0:]], figsize=(12, 12))
````

#### Pandas, Bar Plots

- Use bar plots to compare the percentages of women (ShareWomen) from the 10 highest paying majors and from the 10 lowest paying majors.
